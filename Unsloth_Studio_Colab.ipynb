```python
import subprocess, time, requests, threading, os

# Fix: unset Colab's matplotlib backend before starting Studio
env = os.environ.copy()
env.pop("MPLBACKEND", None)

proc = subprocess.Popen(
    ["/root/.local/bin/unsloth", "studio", "-H", "0.0.0.0", "-p", "8000"],
    stdout=subprocess.PIPE, stderr=subprocess.STDOUT,
    env=env  # pass clean env
)

def stream_logs():
    for line in proc.stdout:
        print(line.decode(), end="")

threading.Thread(target=stream_logs, daemon=True).start()

print("‚è≥ Waiting for Studio to start", end="")
for i in range(90):
    try:
        if requests.get("http://localhost:8000", timeout=2).status_code < 500:
            print(f"\n‚úÖ Studio is up!")
            break
    except:
        print(".", end="", flush=True)
        time.sleep(3)
else:
    print("\n‚ùå Timed out ‚Äî check logs above")
```
